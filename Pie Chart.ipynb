Pie Chart (Proportions of a whole)

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Import 3D axes
import numpy as np
import random
```

```python


# Sample Data (Proportional data)
labels = ['Part 1', 'Part 2', 'Part 3', 'Part 4']
sizes = [3, 25, 15, 30] # Percentages or proportions
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0, 0.1, 0, 2)  # "explode" the 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90) # autopct for % formatting
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Pie Chart Example")
plt.show()
```

Scenario: Imagine you have a fruit basket and you want to visualize the proportion of each type of fruit in it. Let's say you counted the fruits and found the following:

Apples: 5

Bananas: 8

Oranges: 6

Grapes: 11

```python
# Data for the fruit basket
fruit_types = ['Apples', 'Bananas', 'Oranges', 'Grapes']
fruit_counts = [5, 8, 6, 11]
# Colors for each fruit slice (optional, but makes it visually appealing)
colors = ['red', 'yellow', 'orange', 'purple']

# "Explode" the largest slice (Grapes) slightly to make it stand out (optional)
explode = (0.1, 0.21, 0.21, 0.1)  # only "explode" the 4th slice (Grapes)

# Create the pie chart
plt.pie(fruit_counts, explode=explode, labels=fruit_types, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

plt.title("Fruit Composition of Basket")
plt.show()
```
