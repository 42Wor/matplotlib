2.Scatter Plot (Relationship between two variables)

Useful for showing the relationship or correlation between two sets of data.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Import 3D axes
import numpy as np
import random
import pandas as pd
```

```python
# Sample Data (Medium Dataset)
np.random.seed(0) # for reproducibility
x = np.random.rand(500) * 10  # 50 random x values between 0 and 10
y = 1.1 * x + np.random.randn(500) * 3 # y with a linear trend + some noise

plt.scatter(x, y, color='blue', alpha=0.7) # alpha for transparency
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.title("Scatter Plot Example")
plt.grid(True)
plt.show()
```

```python
study_hours = [2, 1, 4, 3, 0.5, 5, 2.5, 3.5, 1.5, 4.5]
exam_scores = [65, 50, 80, 75, 40, 90, 70, 85, 60, 95]

# Create the scatter plot
plt.scatter(study_hours, exam_scores, color='green', marker='o', s=50)
# marker='o' makes the points circles, s=50 sets the size of the markers

# Customize the plot
plt.xlabel("Study Hours")
plt.ylabel("Exam Score (%)")
plt.title("Relationship between Study Hours and Exam Scores")
plt.grid(True)

# Display the plot
plt.show()
```

Data: Imagine you've collected data on daily average temperature and the number of ice cream cones sold on those days. The data might look like this (you can create a CSV file with this content):

```python
data=pd.read_csv("Ice Cream.csv")
# 2. Extract Data for Plotting
temperatures = data['Temperature']
sales = data['Revenue']

# 3. Create the Scatter Plot
plt.figure(figsize=(8, 6)) # Adjust figure size
plt.scatter(temperatures, sales, color='coral', marker='o', s=80, alpha=0.7) # Customize markers and color

# 4. Customize the Plot
plt.title('Ice Cream Sales vs. Temperature', fontsize=16)
plt.xlabel('Temperature (C)', fontsize=12)
plt.ylabel('Ice Cream Sales (Units)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

"Favorite Fruits in a Survey"

```python
data = pd.read_csv('fruit.xls')

# 2. Extract Data for Plotting
mass = data['mass']
width = data['width']
fruit_names = data['fruit_name']

# 3. Create the Scatter Plot, Color-coded by fruit name
plt.figure(figsize=(8, 6))

# Use a dictionary to map fruit names to colors for consistency
fruit_colors = {
    'apple': 'red',
    'mandarin': 'orange',
    'orange': 'darkorange',  # Assuming 'orange' is a different type than 'mandarin'
    'lemon': 'yellow'
}

for fruit in fruit_names.unique(): # Loop through unique fruit names
    fruit_data = data[data['fruit_name'] == fruit] # Filter data for each fruit type
    plt.scatter(fruit_data['mass'], fruit_data['width'],
                label=fruit.capitalize(), # Label for legend
                color=fruit_colors.get(fruit, 'blue'), # Get color from dict, default to blue
                s=80, alpha=0.8) # Adjust marker size and transparency

# 4. Customize the Plot
plt.title('Fruit Mass vs. Width, Colored by Fruit Type', fontsize=14)
plt.xlabel('Mass (g)', fontsize=12)
plt.ylabel('Width (cm)', fontsize=12)
plt.legend(title='Fruit Type') # Add a legend to identify fruit colors
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
