Imagine you're tracking the daily closing price of a stock. You want to see the trend of the stock price over a few weeks.

```python

# 1. Simulate Stock Price Data over 30 days
days = np.arange(1, 31)  # Days 1 to 30
np.random.seed(7) # for reproducibility

# Simulate an upward trend with some random daily fluctuations
base_price = 100 # Starting price
daily_change = 1 + np.random.randn(30) * 0.02 # Average 1% daily increase with 2% std deviation fluctuation
stock_prices = base_price * np.cumprod(daily_change) # Cumulative product to simulate price changes over time

# 2. Create the Line Plot
plt.plot(days, stock_prices, marker='.', linestyle='-', color='blue') # '.' markers, solid line

# 3. Customize the Plot
plt.xlabel("Day of the Month")
plt.ylabel("Stock Price (USD)")
plt.title("Stock Price Trend Over 30 Days")
plt.grid(True, linestyle='--', alpha=0.6) # Dashed grid lines, slightly transparent

# 4. Improve x-axis ticks for days (optional)
plt.xticks(days[::5]) # Show x-axis ticks every 5 days (for better readability)

plt.show()
```

4. Histogram (Distribution of a single variable)

Shows the frequency distribution of a single dataset.

```python

# Sample Data (Larger Dataset for distribution visualization)
np.random.seed(1)
data = np.random.randn(1000) # 1000 random numbers from a normal distribution

plt.hist(data, bins=30, color='skyblue', edgecolor='black') # bins for number of bars
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.grid(axis='y', alpha=0.5) # Grid only on y-axis
plt.show()
```

5. Pie Chart (Proportions of a whole)

```python


# Sample Data (Proportional data)
labels = ['Part 1', 'Part 2', 'Part 3', 'Part 4']
sizes = [30, 25, 15, 30] # Percentages or proportions
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0.1, 0, 0, 0)  # "explode" the 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90) # autopct for % formatting
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Pie Chart Example")
plt.show()
```

6. Box Plot (Statistical Distribution and Outliers)

```python


# Sample Data (Multiple groups)
np.random.seed(2)
data_group1 = np.random.normal(loc=0, scale=1, size=100) # Normal distribution, mean 0, std dev 1
data_group2 = np.random.normal(loc=2, scale=1.5, size=120)
data_group3 = np.random.normal(loc=-1, scale=0.8, size=90)
data = [data_group1, data_group2, data_group3] # List of datasets

labels = ['Group 1', 'Group 2', 'Group 3']

plt.boxplot(data, labels=labels) # Create box plot with labels
plt.ylabel("Values")
plt.title("Box Plot Example")
plt.grid(axis='y', linestyle='--') # Grid on y-axis, dashed lines
plt.show()
```

Imagine you are a teacher and you taught three different classes (Class A, Class B, and Class C). You want to compare how the students performed on the same exam. Box plots are perfect for this!

```python

# 1. Sample Data: Exam Scores for three classes (out of 100)
np.random.seed(0)  # for reproducible results

class_a_scores = np.random.normal(loc=75, scale=10, size=30)  # Class A: Average around 75, some variation
class_b_scores = np.random.normal(loc=82, scale=8, size=35)   # Class B: Average around 82, less variation
class_c_scores = np.random.normal(loc=78, scale=12, size=28)  # Class C: Average around 78, more variation

# Let's also add a potential outlier in Class A (someone who really struggled)
class_a_scores = np.append(class_a_scores, 45) # Add a score of 45 to Class A

exam_scores_data = [class_a_scores, class_b_scores, class_c_scores]
class_labels = ['Class A', 'Class B', 'Class C']

# 2. Create the Box Plot
plt.figure(figsize=(8, 6)) # Adjust figure size for better viewing
plt.boxplot(exam_scores_data, labels=class_labels)

# 3. Customize the Plot
plt.title('Exam Score Comparison Across Classes')
plt.ylabel('Exam Scores (out of 100)')
plt.xlabel('Classes')
plt.grid(axis='y', linestyle='--') # Add a horizontal grid for easier reading

# 4. Show the Plot
plt.show()
```

7. Violin Plot (Distribution Shape and Density)

Violin plots are similar to box plots but show the probability density of the data at different values, giving a more detailed picture of the distribution's shape.

```python

# Sample Data (Same groups as box plot)
np.random.seed(3)
data_group1 = np.random.normal(loc=0, scale=1, size=100)
data_group2 = np.random.normal(loc=2, scale=1.5, size=120)
data_group3 = np.random.normal(loc=-1, scale=0.8, size=90)
data = [data_group1, data_group2, data_group3]
labels = ['Group 1', 'Group 2', 'Group 3']

plt.violinplot(data, showmeans=True, showmedians=False) # Violin plot
plt.xticks(ticks=range(1, len(labels) + 1), labels=labels) # Set x-axis labels
plt.ylabel("Values")
plt.title("Violin Plot Example")
plt.grid(axis='y', linestyle='--')
plt.show()
```

Scenario: Imagine we have exam scores for 5 students in 4 subjects: Math, Science, English, and History. We want to see if students who score well in one subject tend to score well in another (positive correlation), or if there's no relationship, or even a negative correlation (unlikely in this scenario, but possible in other datasets).

```python

# 1. Sample Data: Exam Scores (Students vs. Subjects)
# Rows are students, columns are subjects (Math, Science, English, History)
exam_scores = np.array([
    [85, 92, 78, 88],  # Student 1
    [90, 88, 94, 91],  # Student 2
    [75, 80, 82, 70],  # Student 3
    [68, 72, 70, 65],  # Student 4
    [95, 98, 90, 96]   # Student 5
])

subjects = ["Math", "Science", "English", "History"]

# 2. Calculate the Correlation Matrix
# We want to see the correlation between SUBJECTS, so we transpose the data first
correlation_matrix = np.corrcoef(exam_scores.T)

# 3. Create the Heatmap
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.imshow(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1)
# 'coolwarm' colormap is good for correlations (red=positive, blue=negative, white=no correlation)
# vmin and vmax set the color scale range from -1 to 1 (correlation range)

# 4. Add Colorbar
plt.colorbar(label='Correlation Coefficient')

# 5. Set Ticks and Labels for Subjects
plt.xticks(ticks=range(len(subjects)), labels=subjects, rotation=45, ha="right")
# ticks=range(len(subjects)): places ticks at 0, 1, 2, 3... (indices of subjects)
# labels=subjects: uses subject names as labels
# rotation=45, ha="right": rotates labels for better readability
plt.yticks(ticks=range(len(subjects)), labels=subjects)

# 6. Title
plt.title('Subject Correlation Heatmap of Exam Scores')

# 7. Ensure Labels Fit and Display
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

9. 3D Plots (Visualizing 3D Data)

Matplotlib can also create 3D plots, although they can sometimes be less clear than 2D plots for complex data. You need to import mpl_toolkits.mplot3d.

```python



# Sample Data (3D Surface)
fig = plt.figure() # Create a figure
ax = fig.add_subplot(projection='3d') # Add a 3D subplot

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y) # Create meshgrid for surface plot
Z = np.sin(np.sqrt(X**2 + Y**2)) # Example function for Z values

ax.plot_surface(X, Y, Z, cmap='viridis') # Create surface plot
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot Example')
plt.show()
```

Imagine you're mapping trees in a small rectangular area. You can measure:

    X-coordinate: Position along the east-west direction.

    Y-coordinate: Position along the north-south direction.

    Z-coordinate: Height of the tree.

```python

# 1. Simulate Tree Data
np.random.seed(5) # for reproducibility

num_trees = 50

# X, Y: Positions in a 10x10 meter forest area
x_positions = np.random.rand(num_trees) * 10  # Random x positions (0 to 10 meters)
y_positions = np.random.rand(num_trees) * 10  # Random y positions (0 to 10 meters)

# Z: Tree Heights (let's say heights range from 2 to 15 meters)
tree_heights = 2 + np.random.rand(num_trees) * 13 # Heights between 2 and 15

# 2. Create the 3D Plot
fig = plt.figure(figsize=(8, 6)) # Adjust figure size for better view
ax = fig.add_subplot(projection='3d')

# Scatter plot - each point represents a tree
ax.scatter(x_positions, y_positions, tree_heights, marker='^', s=50, c='green') # '^' for tree-like marker, s for size, c for color

# 3. Customize the plot
ax.set_xlabel('East-West Position (meters)')
ax.set_ylabel('North-South Position (meters)')
ax.set_zlabel('Tree Height (meters)')
ax.set_title('Locations and Heights of Trees in a Forest')

# Set axis limits to match the data range (optional, but good for clarity)
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.set_zlim([0, 15]) # Extend z limit a bit to show heights clearly

# Optionally, adjust viewing angle (experiment with these values)
ax.view_init(elev=25, azim=-45) # elev: elevation angle, azim: azimuth angle

plt.show()
```

10. Subplots (Combining Multiple Plots in One Figure)

You can arrange multiple plots within a single figure using subplots. This is useful for comparing different visualizations side-by-side.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure with 2 subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(10, 5)) # figsize for figure size

# Plot on the first subplot (axes[0])
axes[0].plot(x, y1, color='blue')
axes[0].set_title('Sine Wave')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

# Plot on the second subplot (axes[1])
axes[1].scatter(x, y2, color='red', marker='x')
axes[1].set_title('Cosine Wave (Scatter)')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')

plt.tight_layout() # Adjust subplot parameters for a tight layout
plt.suptitle("Combined Subplots Example") # Figure-level title
plt.show()
```

Imagine you're tracking the weather for a single day and want to visualize two key variables: temperature and rainfall over time. Using subplots, we can display both plots in one figure for easy comparison. 

```python

# 1. Simulate Weather Data for a Day (24 hours)
hours = np.arange(24)  # Hours of the day (0 to 23)

# Temperature Data (in Celsius) - Let's simulate a warm day with some variation
np.random.seed(6) # for reproducibility
temperature = 20 + 8 * np.sin(2 * np.pi * hours / 24) + 2 * np.random.randn(24)
# Base temperature 20C, sinusoidal variation (daily cycle), and some random noise

# Rainfall Data (in mm) - Let's simulate a short rain shower in the afternoon
rainfall = np.zeros(24) # Initialize rainfall to zero for all hours
rainfall[14:17] = [1, 2, 1.5] # Rain shower from 2 PM to 5 PM (hours 14, 15, 16)

# 2. Create Subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(12, 5)) # Wider figure to accommodate two plots

# 3. Plot Temperature on the first subplot (axes[0])
axes[0].plot(hours, temperature, color='red', linestyle='-', marker='o')
axes[0].set_title('Temperature (C) over the Day')
axes[0].set_xlabel('Hour of the Day (0-23)')
axes[0].set_ylabel('Temperature (C)')
axes[0].grid(True, linestyle=':', alpha=0.7) # Add a light grid

# 4. Plot Rainfall on the second subplot (axes[1])
axes[1].bar(hours, rainfall, color='skyblue') # Bar chart for rainfall
axes[1].set_title('Rainfall (mm) over the Day')
axes[1].set_xlabel('Hour of the Day (0-23)')
axes[1].set_ylabel('Rainfall (mm)')
axes[1].grid(axis='y', linestyle=':', alpha=0.7) # Grid only on y-axis

# 5. Adjust layout and add a figure title
plt.tight_layout() # Improve subplot spacing
plt.suptitle('Weather Data for a Day', fontsize=16) # Overall figure title, larger font

plt.show()
```
