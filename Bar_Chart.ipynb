3. Bar Chart (Comparison of Categories)

Ideal for comparing values across different categories or groups.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Import 3D axes
import numpy as np
import random
import pandas as pd
```

```python

categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 40, 30, 55, 15]

plt.bar(categories, values, color=['red', 'green', 'blue', 'orange', 'purple']) # Different colors
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart Example")
plt.show()
```

Imagine you are managing a small fruit stand, and you want to visualize the sales of different types of fruits you sold in a week to see which ones are most popular.

```python

fruit_types = ['Apple', 'Banana', 'Orange', 'Grape', 'Strawberry']
fruit_sales = [150, 200, 120, 80, 90]

# Create the bar chart
plt.bar(fruit_types, fruit_sales, color=['red', 'yellow', 'orange', 'purple', 'pink'])
# You can specify colors for each bar for better visualization

# Customize the plot
plt.xlabel("Fruit Type")
plt.ylabel("Number of Fruits Sold")
plt.title("Weekly Fruit Sales at Fruit Stand")
plt.grid(axis='y', linestyle='--', alpha=0.7) # Add a horizontal grid for easier comparison

# Display the plot
plt.show()
```

 "Favorite Fruits in a Survey"

```python
# 1. Load the Data from CSV (or Excel if you're using 'fruit.xls' and pd.read_excel)
data = pd.read_csv('fruit.xls')  # Or data = pd.read_excel('fruit.xls') if using Excel

# 2. Count the occurrences of each fruit name
fruit_counts = data['fruit_name'].value_counts()
# value_counts() gives you a Series with unique values as index and their counts as values.

# 3. Extract fruit names and their counts for the bar chart
fruit_names = fruit_counts.index  # Fruit names will be the index of the Series
counts = fruit_counts.values     # Counts will be the values of the Series

# 4. Create the Bar Chart
plt.figure(figsize=(8, 6)) # Adjust figure size if needed
plt.bar(fruit_names, counts, color='skyblue') # Basic bar chart, you can customize colors

# 5. Customize the Plot
plt.title('Number of Fruits by Type in the Dataset', fontsize=14)
plt.xlabel('Fruit Type', fontsize=12)
plt.ylabel('Number of Fruits', fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for readability if needed
plt.grid(axis='y', linestyle='--', alpha=0.7) # Add a y-axis grid for better readability

plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```
