Imagine you're tracking the daily closing price of a stock. You want to see the trend of the stock price over a few weeks.

```python
# 1. Simulate Stock Price Data over 30 days
days = np.arange(1, 31)  # Days 1 to 30
np.random.seed(7) # for reproducibility

# Simulate an upward trend with some random daily fluctuations
base_price = 100 # Starting price
daily_change = 1 + np.random.randn(30) * 0.02 # Average 1% daily increase with 2% std deviation fluctuation
stock_prices = base_price * np.cumprod(daily_change) # Cumulative product to simulate price changes over time

# 2. Create the Line Plot
plt.plot(days, stock_prices, marker='.', linestyle='-', color='blue') # '.' markers, solid line

# 3. Customize the Plot
plt.xlabel("Day of the Month")
plt.ylabel("Stock Price (USD)")
plt.title("Stock Price Trend Over 30 Days")
plt.grid(True, linestyle='--', alpha=0.6) # Dashed grid lines, slightly transparent

# 4. Improve x-axis ticks for days (optional)
plt.xticks(days[::5]) # Show x-axis ticks every 5 days (for better readability)

plt.show()
```

```python
csv_filename = 'website_traffic.csv'  # Name of the CSV file to create
num_days = 100                       # Number of days of data
start_date = datetime.date(2023, 1, 1)  # Starting date for the data
min_visitors = 500                    # Minimum daily visitors
max_visitors = 3000                   # Maximum daily visitors

dates = [start_date + datetime.timedelta(days=i) for i in range(num_days)]
dates_str = [date.strftime('%Y-%m-%d') for date in dates]
np.random.seed(42)  
visitors = np.random.randint(min_visitors, max_visitors + 1, size=num_days)
data = {'Date': dates_str, 'Visitors': visitors}
df = pd.DataFrame(data)

df.to_csv(csv_filename, index=False)  # index=False prevents writing row indices to the CSV

print(f"CSV file '{csv_filename}' created successfully with {num_days} days of random website traffic data.")
```

```python
# --- 2. Extract Data for Plotting ---
data = pd.read_csv('website_traffic.csv') # Load the CSV file 100 rows


dates = pd.to_datetime(data['Date']) # Convert 'Date' column to datetime objects
visitors = data['Visitors']

# --- 3. Create and Customize the Line Plot ---
plt.figure(figsize=(12, 6)) # Figure size for better viewing - reduced from (80, 50) to a more reasonable size
plt.plot(dates, visitors, marker='o', linestyle='-', color='purple')

plt.title('Website Traffic Over Time', fontsize=14) # Changed title to be more general and accurate for daily data
plt.xlabel('Date', fontsize=10)
plt.ylabel('Number of Visitors', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.6)

# Format x-axis to display dates nicely
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d')) # Format dates on x-axis
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator()) # Automatically choose date intervals

plt.xticks(rotation=45, ha='right') # Rotate date labels - keep this for readability

plt.tight_layout()

# --- 4. Show the Plot ---
plt.show()
```
