Imagine you're tracking the daily closing price of a stock. You want to see the trend of the stock price over a few weeks.

```python
# 1. Simulate Stock Price Data over 30 days
days = np.arange(1, 31)  # Days 1 to 30
np.random.seed(7) # for reproducibility

# Simulate an upward trend with some random daily fluctuations
base_price = 100 # Starting price
daily_change = 1 + np.random.randn(30) * 0.02 # Average 1% daily increase with 2% std deviation fluctuation
stock_prices = base_price * np.cumprod(daily_change) # Cumulative product to simulate price changes over time

# 2. Create the Line Plot
plt.plot(days, stock_prices, marker='.', linestyle='-', color='blue') # '.' markers, solid line

# 3. Customize the Plot
plt.xlabel("Day of the Month")
plt.ylabel("Stock Price (USD)")
plt.title("Stock Price Trend Over 30 Days")
plt.grid(True, linestyle='--', alpha=0.6) # Dashed grid lines, slightly transparent

# 4. Improve x-axis ticks for days (optional)
plt.xticks(days[::5]) # Show x-axis ticks every 5 days (for better readability)

plt.show()
```
